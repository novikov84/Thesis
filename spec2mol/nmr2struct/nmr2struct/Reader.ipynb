```python
import json
import pandas as pd
```

```python
import polars as pl
from tqdm import tqdm
```

```python
def get_input_vect(signals):
    #на выходе вектор длиной 300, каждый элемент которого = количество сигналов в конкретном бте  
    #от -50 до 250 включительно, каждый элемент которого = интенсивность
    result = [0]*300
    for signal in signals:
        position = int(signal- (-50))
        result[position] += 1
    
    return result
```

```python
list_columns = ['input', 'fp2frags', 'fp3frags', 'fp4frags',  'signals']
```

```python
df = pl.read_csv('data/spec_df.csv', separator='\t', truncate_ragged_lines=True)
```

```python
df.shape[0]
```

```python
processed_data = []

for idx in tqdm(range(df.shape[0])):
    processed_data.append({
        'spectrum' : json.loads(df['input'][idx]),
        'frags' : json.loads(df['fp2frags'][idx]) 
    })
```

```python
json.dump(processed_data, open('processed_spec_data.json', 'w'))
```

```python
json.dump(processed_data[:100], open('processed_spec_data_sample.json', 'w'))
```
